```python
# Step 1: Drop rows where 'demand' is missing (essential target variable)
combined_df = combined_df.dropna(subset=['demand'])

# Step 2: Impute missing weather features with median (low missing %)
weather_features_to_impute = [
    'temperature', 'apparentTemperature', 'humidity',
    'pressure', 'windSpeed', 'windBearing', 'windGust',
    'cloudCover', 'uvIndex', 'visibility'
]
combined_df[weather_features_to_impute] = combined_df[weather_features_to_impute].fillna(
    combined_df[weather_features_to_impute].median()
)
```

```python
combined_df = combined_df.drop(columns=['precipType', 'precipAccumulation', 'ozone'], errors='ignore')
```

```python
# Step 4: Fill missing text fields with 'Unknown'
combined_df['summary'] = combined_df['summary'].fillna('Unknown')
combined_df['icon'] = combined_df['icon'].fillna('Unknown')
```

```python
# --- 2. Feature Engineering ---
# Convert 'local_time' column to datetime
combined_df['local_time'] = pd.to_datetime(combined_df['local_time'])
```

```python
# Extract time-based features: hour, day of week, month, season
combined_df['hour'] = combined_df['local_time'].dt.hour
combined_df['day_of_week'] = combined_df['local_time'].dt.dayofweek
combined_df['month'] = combined_df['local_time'].dt.month
```

```python
# Define season based on month
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

combined_df['season'] = combined_df['month'].apply(get_season)
```

```python
# --- 3. Normalize/Scale Continuous Variables ---
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
continuous_features = ['demand', 'temperature', 'apparentTemperature', 'humidity', 'windSpeed', 'pressure']
combined_df[continuous_features] = scaler.fit_transform(combined_df[continuous_features])
```

```python
# --- 4. Aggregation: Compute Daily Summary Statistics ---
combined_df['date'] = combined_df['local_time'].dt.date
```

```python
# Compute daily summary statistics for demand and weather features
daily_summary = combined_df.groupby('date').agg({
    'demand': ['mean', 'max', 'min'],
    'temperature': ['mean', 'max', 'min'],
    'humidity': 'mean',
}).reset_index()

daily_summary.columns = ['date', 'demand_mean', 'demand_max', 'demand_min',
                         'temp_mean', 'temp_max', 'temp_min', 'humidity_mean']
```

```python
from scipy.stats import zscore
```

```python
# ------------------------ 5. ANOMALY DETECTION ------------------------
# Z-Score for demand
combined_df['demand_z'] = zscore(combined_df['demand'])
anomalies_zscore = combined_df[combined_df['demand_z'].abs() > 3]
```

```python
# ----- Elbow Method for KMeans -----
inertias = []
K = range(2, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(pca_result)
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(6, 4))
plt.plot(K, inertias, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia (WCSS)')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```
